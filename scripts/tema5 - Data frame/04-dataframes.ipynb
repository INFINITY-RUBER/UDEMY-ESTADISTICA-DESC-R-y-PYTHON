# CÃ³mo crear un data frame

```python
import numpy as np # cargamos la libreria analices numerico y matematica de python
```

```python
df = pd.DataFrame({
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10)
})
```

```python
df
```

```python
df.mean() # media
```

```python
df.mean(axis='columns') # media por columnas
```

```python
df2 = pd.DataFrame({
    'key' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'values' : range(6)
},
columns = ['key', 'values']
)
df2 # crear un data frame con dicionarios manualmente
```

## agregados para operaciones

```python
df2.groupby('key')# agrupar por el mismo clave.  es un objeto 
```

```python
df2.groupby('key').mean() # agrupar por clave y saca la media aritmetica
```

```python
df2.groupby('key').median() # agrupar por clave y saca la mediana
```

# ManipulaciÃ³n de data frames

```python
df = pd.DataFrame({
       'key' : ['A', 'B', 'C', 'A', 'B', 'C'],
       'value1': range(6),
       'value2': np.random.randint(0, 100, 6) # 6 numeros aleatorios entre 1 - 100
})
```

```python
df
```

## Aggregate

```python
df.groupby('key').aggregate(['min', np.median, np.mean, max])# agrega a cada valor [min  median  mean max]
```

```python
df.groupby('key').aggregate({
    'value1': 'min',
    'value2': 'max'
}) # agrega a cada valor segun el dicionario
```

## Filter

```python
df.groupby('key').std() # agrupa con la Desviacion estandar
```

```python
def filter_func(x):
    return x['value2'].std()>5
```

```python
display('df', "df.groupby('key').std()", "df.groupby('key').filter(filter_func)")
```

```python
df.groupby('key').filter(filter_func)
```

## Transform

```python
df.groupby('key').transform(lambda x: x - x.mean())# transforma el df con con su valor - media 
```

## Apply  
(aplicar)

```python
def norm_by_col2(x):
    x['value1'] /= x['value2'].sum() #x['value1'] = x['value1'] / x['value2'].sum()
    return x
```

```python
df.groupby('key').apply(norm_by_col2)
```

# GrÃ¡ficos desde data frames y exportaciÃ³n

```python
import numpy as np
years = [year for year in range(1900, 2020)]
deads = [(y + np.random.uniform(0,100) - 1850) for y in years]
```

```python
df = pd.DataFrame({
    "year": years,
    "deads": deads
})
```

```python
df.head()
```

```python
df.plot( x="year", y = "deads", figsize = (10, 8))
```

```python
my_plot = df.plot(kind = "scatter", x="year", y = "deads", figsize = (10, 8)) # creamos la variable de la imagen
```

```python
my_fig = my_plot.get_figure() # asignamos la figura a una variable my_fig
```

```python
my_fig.savefig("../../ejercicios/data_frame_plot2.png") # guardamos la imagen creada
```
