## libreria Matplotlib 

```python
%matplotlib inline 
# para llamar la libreria y trabaje en lina igual a R
```

```python
import matplotlib.pyplot as plt  # le colocamos un Alias
```

```python
x = [1,2,3,4]
```

```python
plt.plot(x) # grafica y colocar texto 
plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de ordenadas")
plt.show() # para configurar y unifique todo 
```

```python
x = [1,2,3,4] 
y = [1,4,9,16]
plt.plot(x, y)# pintar las dos funciones
```

```python
plt.plot(x,y,"ro")  # "ro" significa que 'r' de Rojo y 'o' de cierculos
plt.axis([0, 6, 0, 20]) # amplio el rango del plano .axis([iniX, finX, iniY, finY])
plt.show() # cierro 
```

```python
import numpy as np # importamos numpy libreria de calculos matematicos
data = np.arange(0.0, 10.0, 0.2)
data
```

```python
plt.plot(data, data, "r--", # "r--" significa que 'r' de Rojo y '--' de guiones
         data, data**2, "bs", # "bs" significa que 'b' de blue y 's' de cuadros
         data, data**3, "g^") # "g^" significa que 'g' de green y '^' de triangulos
plt.show()
```
