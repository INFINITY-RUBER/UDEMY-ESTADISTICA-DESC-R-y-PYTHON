1- **A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 1 en el enunciado. Pista: investiga el parámetro fill_between de matplotlib.**

n = 256

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.plot(X, Y + 1, color='blue', alpha=1.00)# grafica N1
ax=plt.gca()
ax.fill_between(X, 1, Y+1, color='#9999ff')

plt.plot(X, Y - 1, color='blue', alpha=1.00)# grafica N2
ax=plt.gca()
ax.fill_between(X, -1, Y-1, color='#9999ff', where=((Y-1) >= -1))
ax.fill_between(X, -1, Y-1, color='#ff9999',    where=((Y-1) <= -1))
plt.xticks([],[])# sin marcas en x ([],[])
plt.yticks([],[]) # sin marcas en y ([],[])
plt.xlim(-np.pi, np.pi)
plt.ylim(-2.5, 2.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

n = 12

X = np.arange(n)

Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')

plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):

    plt.text(x + 0.1, y + 0.05, '%.2f' % y, ha='center', va='bottom',fontsize = 14) 
 
for x, y in zip(X, Y2):

    plt.text(x + 0.1, -y -0.05, '%.2f' % y, ha='center', va='top',fontsize = 14) 
    
plt.xlim(-.5, n)
plt.ylim(-1.25, 1.25) # limite eje y
plt.xticks([],[])# sin marcas en x ([],[])
plt.yticks([],[]) # sin marcas en y ([],[])
plt.show()
```

```python
plt.figure(figsize=(16,10))

n = 20

Z = np.ones(n)

Z[-1] *= 2

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.pie(Z, explode=Z*.06, colors = ['%f' % (i/float(n)) for i in range(n)])

plt.axis('equal')
plt.xticks([],[])# sin marcas en x ([],[])
plt.yticks([],[]) # sin marcas en y ([],[])
plt.show()
```
