## M√∫ltiples gr√°ficos en una misma figura con libreria numpy

```python
import numpy as np # importamos numpy libreria de calculos matematicos
```

```python
#creamaos una funcion para reutilizarla
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
# creamos los vectores de datos para que pinte la grafica
x1 = np.arange(0, 5.0, 0.1) # np.arange(inicio, fin, saltos)
x2 = np.arange(0, 5.0, 0.2)
```

```python
plt.figure(1)
plt.subplot(2,1,1) # tambien se puede usar  plt.subplot(211)
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k') # por un lado pinta los puntos y por otro lineas que une los puntos
```

```python
plt.figure(1)
plt.subplot(2,1,1) # tambien se puede usar  plt.subplot(211) (el doble que de ancho)
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k')

plt.subplot(2,1,2) # subplot(2,1,2) quiere indicar el numero de filas y columnas a indicar
plt.plot(x2, f(x2), 'g-.')

plt.show()
```

# Textos en los gr√°ficos

```python
mu = 100 # media 
sigma = 20 # desviacion tipica

x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="b", alpha=0.6)

plt.xlabel("Cociente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu,\sigma)$")
plt.text(120, 0.015, r'$\mu = 100,\ \sigma=20$')
plt.axis([20,180,0, 0.025])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6), dpi = 90) # figsize=(Tama√±o grafica); dpi =(pixel por pugada)
plt.subplot(1,1,1)

x = np.arange(0, 8*np.pi, 0.01)
y = np.cos(x)
plt.plot(x,y, lw = 2.0)

plt.annotate('M√°ximo Local', xy = (4*np.pi, 1), xytext = (15, 1.5), # anotacion en grafica
            arrowprops = dict(facecolor = "red", shrink = 0.5))
plt.ylim(-2,2)
plt.show()
```

## cambios de ejes y escalas

```python
from matplotlib.ticker import NullFormatter # libreria para formatos  de estilos
```

```python
mu = 0.5 # media
sd = 0.3  # desviacion
y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)] # filtro los datos
y.sort()
x = np.arange(len(y))
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)# grafica 1
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(2,2,2)# grafica 2
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala Logar√≠tmica")
plt.grid(True)

plt.subplot(2,2,3) # grafica 3
plt.plot(x, y - y.mean())
plt.yscale("symlog", linthreshy=0.01)
plt.title("Escala Log Sim√©trico")
plt.grid(True)

plt.subplot(2,2,4) # grafica 4
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala log√≠stica")
# plt.gca() = grafica actual
plt.gca().yaxis.set_minor_formatter(NullFormatter()) # para eliminar todo formato de las Ejes de y
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35)
# hspace = ALTURA; wspace = ANCHO
plt.show()
```

# Cambios en los ejes de dibujo

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True) # si no se coloca endpoint=True  no coje el ultimo digito
S, C = np.sin(x), np.cos(x) # asignar dos fucniones separadas por la coma 
```

```python
plt.figure(figsize=(10,8))

plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-.", label = "seno")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-", label = "coseno")
plt.xlim(x.min()*1.1, x.max()*1.1) # tomo el de x el valor minimo y maximo y se multiplica por 10% 
plt.ylim(S.min()*1.1, S.max()*1.1) # EL LIMITE DE Y
# Las marcas en el eje se llaman ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$'])
plt.yticks(np.linspace(-1,1, 2, endpoint=True),
          ['-1',  '+1'])

ax = plt.gca() # get curren access 'aceder ala grafica actual'
ax.spines['right'].set_color('none')# quita barra de la derecha
ax.spines['top'].set_color('none')# quita barra de la arriba
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))# Mueve la barra x de abajo a posicion 0
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))# Mueve la barra y de la izquierda a posicion 0

plt.legend(loc="upper left",fontsize = 16)# leyenda de etiquetas

x0 = 2*np.pi/3
# pintar una linea en de intercesion en 2*pi/3; linewidth = grosor de la linea
plt.plot([x0,x0], [0, np.sin(x0)], color = "blue", linewidth = 3.5, linestyle = "--")
plt.scatter([x0, ], [np.sin(x0), ], 100, color = "red") # pinta un punto rojo en (0,sen(2*pi/3))

# anotacion del punto en Markdown seno
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$', 
            xy = (x0, np.sin(x0)), xycoords = "data",
            xytext = (+10,+80), textcoords = "offset points",# (+10,+30)del punto 
            fontsize = 16, # tama√±o del texto
            arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.10"))# flecha al texto

plt.plot([x0,x0], [0, np.cos(x0)], color = "red", linewidth = 3.5, linestyle = "--")
plt.scatter([x0, ], [np.cos(x0), ], 100, color = "green") # pinta un punto rojo en (0,sen(2*pi/3))
# anotacion del punto en Markdown coseno
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$', 
            xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (-90,-60), textcoords = "offset points",# (+10,+30)del punto 
            fontsize = 16, # tama√±o del texto
            arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.10"))# flecha al texto
# resaltar las etiquetas numericas
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='green', edgecolor='None', alpha = 0.6))

plt.show()
```
